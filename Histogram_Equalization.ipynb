```
import cv2
from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt
import numpy as np

img = cv2.imread('kidney.tif',0)
cv2_imshow(img)

rows = img.shape[0]
cols = img.shape[1]

#Generating a mask for intensity values greater than 145
mask = img > 145  # vectorization 

#Now applying that mask on Image to give those intensities max value of 255
img[mask] = 255 # vectorization 

cv2_imshow(img)


    
```

```
#Image Reading
img1 = cv2.imread('wiki.jpg')
cv2_imshow(img1)

#Basic Calculations from image 
rmax = np.max(img1)
rmin = np.min(img1)
smax = 255
smin = 0

#Formula
s = (smax - smin)/(rmax - rmin)*(img1 - rmin) + smax
cv2_imshow(s)

#Second Image
img1 = cv2.imread('lowcon.tif')
cv2_imshow(img1)

rmax = np.max(img1)
rmin = np.min(img1)
smax = 255
smin = 0

s = (smax - smin)/(rmax - rmin)*(img1 - rmin) + smax
cv2_imshow(s)
```

```
img1 = cv2.imread('dark.tif',0)
cv2_imshow(img1)

list1 = np.zeros(256)
output = np.zeros(256)

rows = img1.shape[0]
cols = img1.shape[1]


#Frequency distribution of pixels in Image 
for y in range(cols):
  for x in range(rows):
    intensity = img1[x][y]
    list1[intensity] += 1 #counting

#Comulative Frequency Distribution
output[0] = list1[0]
for i in range(1,len(list1)):
    output[i] = output[i-1] + list1[i]

#Normalization suing enumeration
output = [ele*255/output[-1] for ele in output]

#Image equalization using interpolation
histogram_Equalized = np.interp(img1, range(0,256), output)
cv2_imshow(histogram_Equalized)

equ = cv2.equalizeHist(img1) #Using Opencv Built in method
cv2_imshow(equ)
```
